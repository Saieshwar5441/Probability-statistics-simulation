## **4.1.3 Variance Reduction Techniques - Conditional Monte Carlo**

```python
import random

def roll_die():
    return random.randint(1, 6)

def monte_carlo_simulation(num_simulations):
    outcomes = [0] * 6
    for _ in range(num_simulations):
        outcome = roll_die()
        outcomes[outcome - 1] += 1
    return outcomes

def conditional_monte_carlo_simulation(num_simulations, exclude_outcome):
    outcomes = [0] * 6
    for _ in range(num_simulations):
        outcome = roll_die()
        if outcome != exclude_outcome:
            outcomes[outcome - 1] += 1
    return outcomes


num_simulations = 10000

print("Monte Carlo Simulation Results:")
results = monte_carlo_simulation(num_simulations)
for outcome, count in enumerate(results, start=1):
    print(f"Outcome {outcome}: {count} occurrences")

exclude_outcome = 3
print("\nConditional Monte Carlo Simulation Results:")
results_conditional = conditional_monte_carlo_simulation(num_simulations, exclude_outcome)
for outcome, count in enumerate(results_conditional, start=1):
    print(f"Outcome {outcome}: {count} occurrences (excluding {exclude_outcome})")
```

## **4.1.4 Markov Chains**

## **Transition Matrix Simulation**

```python
import numpy as np

def simulate_markov_chain(transition_matrix, initial_state, steps):
    num_states = len(transition_matrix)
    state_sequence = [initial_state]

    for _ in range(steps):
        current_state = state_sequence[-1]
        next_state = np.random.choice(np.arange(num_states), p=transition_matrix[current_state])
        state_sequence.append(next_state)

    return state_sequence

def calculate_state_probabilities(state_sequence):
    num_states = len(set(state_sequence))
    state_counts = np.bincount(state_sequence, minlength=num_states)
    state_probabilities = state_counts / len(state_sequence)
    return state_probabilities

transition_matrix = np.array([[0.7, 0.2, 0.1],
                              [0.3, 0.5, 0.2],
                              [0.1, 0.4, 0.5]])

initial_state = 0
steps = 100

state_sequence = simulate_markov_chain(transition_matrix, initial_state, steps)
state_probabilities = calculate_state_probabilities(state_sequence)

print(f"Simulated State Sequence: {state_sequence}")
print(f"Calculated State Probabilities: {state_probabilities}")
```

## **Ergodicity**

```python
import numpy as np

def ergodicity_simulation(transition_matrix, initial_state, num_steps, num_simulations):
    state_counts = np.zeros(len(transition_matrix))
    time_averaged_probabilities = np.zeros(len(transition_matrix))

    for _ in range(num_simulations):
        current_state = initial_state
        for _ in range(num_steps):
            current_state = np.random.choice(range(len(transition_matrix)), p=transition_matrix[current_state])
            state_counts[current_state] += 1
    time_averaged_probabilities = state_counts / (num_steps * num_simulations)
    return time_averaged_probabilities


num_steps_ergodicity = 100
num_simulations_ergodicity = 1000

ergodicity_result = ergodicity_simulation(transition_matrix_recurrent, initial_state_recurrent, num_steps_ergodicity, num_simulations_ergodicity)
print("Time-Averaged Probabilities:", ergodicity_result)
```

## **Sensitivity Analysis**

```python
import numpy as np
import matplotlib.pyplot as plt

def sensitivity_analysis_simulation(base_transition_matrix, perturbation_factor, num_steps):
    perturbed_transition_matrix = base_transition_matrix * (1 + perturbation_factor * np.random.uniform(size=base_transition_matrix.shape))
    perturbed_transition_matrix /= np.sum(perturbed_transition_matrix, axis=1, keepdims=True)
    state_probabilities = np.zeros(base_transition_matrix.shape[0])
    current_state = np.random.choice(range(len(base_transition_matrix)))
    for _ in range(num_steps):
        current_state = np.random.choice(range(len(perturbed_transition_matrix)), p=perturbed_transition_matrix[current_state])
        state_probabilities[current_state] += 1
    state_probabilities /= num_steps
    return state_probabilities

def plot_sensitivity_analysis(base_transition_matrix, perturbation_factor_range, num_steps, num_simulations):
    results = []
    for perturbation_factor in perturbation_factor_range:
        sensitivity_result = sensitivity_analysis_simulation(base_transition_matrix, perturbation_factor, num_steps)
        results.append(sensitivity_result)
    results = np.array(results).T
    plt.figure(figsize=(10, 6))
    for i in range(len(base_transition_matrix)):
        plt.plot(perturbation_factor_range, results[i], label=f"State {i}")

    plt.title("Sensitivity Analysis: Varying Transition Probabilities")
    plt.xlabel("Perturbation Factor")
    plt.ylabel("State Probability")
    plt.legend()
    plt.show()


transition_matrix = np.array([[0.7, 0.3], [0.8, 0.8]])
perturbation_factor_range = np.linspace(0, 0.5, 10)
num_steps = 100
num_simulations = 100

plot_sensitivity_analysis(transition_matrix, perturbation_factor_range, num_steps, num_simulations)
```

### **4.1.5 Comparision: Estimating Pi value using Monte Carlo Simulation VS Markov Chain Monte Carlo Simulation(MCMC)**

```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_simulation(num_samples):
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    inside_circle = np.sqrt(x**2 + y**2) <= 1
    estimated_pi = 4 * np.sum(inside_circle) / num_samples
    return estimated_pi

def mcmc_simulation(num_samples, num_steps):
    current_position = np.array([1.0, 1.0])

    accepted_samples = 0

    for _ in range(num_samples):
        proposed_position = current_position + np.random.uniform(-0.5, 0.5, 2)
        distance_current = np.sqrt(np.sum(current_position**2))
        distance_proposed = np.sqrt(np.sum(proposed_position**2))
        if distance_proposed <= 1:
            current_position = proposed_position
            accepted_samples += 1
    estimated_pi = 4 * accepted_samples / num_samples
    return estimated_pi

num_samples = 1000
num_mcmc_steps = 1000

monte_carlo_result = monte_carlo_simulation(num_samples)
mcmc_result = mcmc_simulation(num_samples, num_mcmc_steps)

print("Monte Carlo Estimated π:", monte_carlo_result)
print("MCMC Estimated π:", mcmc_result)


x = np.linspace(-1, 1, 100)
y = np.sqrt(1 - x**2)
plt.figure(figsize=(8, 8))
plt.scatter(*np.random.uniform(-1, 1, (2, num_samples)), color='blue', alpha=0.5, label='Monte Carlo Samples')
plt.plot(x, y, 'r-', label='True Circle')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Monte Carlo Simulation for π Estimation')
plt.legend()
plt.show()
```

## **4.2.1 Bayes Theorem: probability of the received email being a spam mail.**

```python
import re
from collections import defaultdict

# Sample email dataset
emails = [
    {'text': 'Get free money!', 'label': 'spam'},
    {'text': 'Hello, how are you?', 'label': 'not spam'}
]

# Function to preprocess text: tokenize words, remove punctuation, and convert to lowercase
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    words = re.findall(r'\b\w+\b', text)  # Tokenize words
    return words

# Initialize dictionaries to store word counts for spam and non-spam emails
spam_word_counts = defaultdict(int)
non_spam_word_counts = defaultdict(int)
spam_total = 0
non_spam_total = 0

for email in emails:
    words = preprocess_text(email['text'])
    label = email['label']

    for word in words:
        if label == 'spam':
            spam_word_counts[word] += 1
            spam_total += 1
        else:
            non_spam_word_counts[word] += 1
            non_spam_total += 1

# Calculate probabilities of words given spam and non-spam
spam_probabilities = {word: spam_word_counts[word] / spam_total for word in spam_word_counts}
non_spam_probabilities = {word: non_spam_word_counts[word] / non_spam_total for word in non_spam_word_counts}

# Function to classify an email as spam or not spam using Bayes' Theorem
def classify_email(text):
    words = preprocess_text(text)

    # Calculate the initial probability of being spam and not spam
    # Assuming equal prior probabilities for spam and non-spam
    prob_spam = prob_not_spam = 0.5

    for word in words:
        # Use Laplace smoothing to handle new words not in the training set
        prob_spam *= (spam_probabilities.get(word, 0) + 1)
        prob_not_spam *= (non_spam_probabilities.get(word, 0) + 1)

    # Apply Bayes' Theorem
    prob_spam_given_email = prob_spam / (prob_spam + prob_not_spam)

    return prob_spam_given_email

# Test the classification for a new email
new_email_text = "Congratulations! You've won a prize!"
classification = classify_email(new_email_text)
# Test the classification for a new email
probability_spam = classify_email(new_email_text)

print(f"The probability of the email being classified as spam is: {probability_spam:.2f}")


if classification > 0.5:
    print("The email is classified as spam.")
else:
    print("The email is classified as not spam.")
```

The probability calculated describes that the email is classified as a spam as it was calculated to be 44% as spam only.

## **4.2.2 Joint Distribution Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/2023_nba_player_stats.csv'
nba_data = pd.read_csv(file_path)

print(nba_data.head())

# Function to calculate conditional probability
def calculate_conditional_probability_nba(data, variable1, threshold1, variable2, threshold2):
    count_both = data[(data[variable1] > threshold1) & (data[variable2] > threshold2)].shape[0]
    count_variable2 = data[data[variable2] > threshold2].shape[0]
    conditional_probability = count_both / count_variable2 if count_variable2 > 0 else 0
    return conditional_probability

# Define variables and their thresholds (median values)
variable1 = 'PTS'  # Points
variable2 = 'AST'  # Assists
threshold1 = nba_data[variable1].median()
threshold2 = nba_data[variable2].median()

# Calculate conditional probability
conditional_probability_nba = calculate_conditional_probability_nba(nba_data, variable1, threshold1, variable2, threshold2)
print(f"Conditional Probability of having more than median {variable1} given more than median {variable2}: {conditional_probability_nba:.2f}")

# Function to check for independence using conditional probabilities
def check_independence_nba(data, variable1, variable2, threshold=0.1):
    threshold1 = data[variable1].median()
    threshold2 = data[variable2].median()

    conditional_prob = calculate_conditional_probability_nba(data, variable1, threshold1, variable2, threshold2)
    marginal_prob_variable1 = (data[data[variable1] > threshold1].shape[0]) / data.shape[0]

    return abs(conditional_prob - marginal_prob_variable1) <= threshold

# Checking independence
independence_result_nba = check_independence_nba(nba_data, variable1, variable2)
print(f"\nAre {variable1} and {variable2} independent? {independence_result_nba}")

# Correlation visualization
correlation_matrix_nba = nba_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_nba, annot=False, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix for NBA Player Stats 2023')
plt.show()
```

Heatmap determines the players performance.For exaple, A dark red square will most likely be present where "PTS" and "FGM" meet, meaning that players who make more field goals will score higher.
